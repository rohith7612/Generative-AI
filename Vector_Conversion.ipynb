# One hot encoding.

```python
from sklearn.preprocessing import OneHotEncoder

sentence = "Taj Mahal is a beautiful monument"
words = sentence.lower().split()

words_reshaped = [[word] for word in words]

encoder = OneHotEncoder(sparse_output=False)
encoded = encoder.fit_transform(words_reshaped)

print(encoded)
```

# Bag of Words technique

```python
from sklearn.feature_extraction.text import CountVectorizer

sentence = ["Taj Mahal is a beautiful monument"]

vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(sentence)

print("Vocabulary:", vectorizer.vocabulary_)
print("Bag of Words Vector:", bow.toarray())
```

# Data Cleaning and Preprocessing.

```python
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
```

```python
corpus = []
review = re.sub('[^a-zA-Z]', ' ', messages)
review = review.lower()
review = review.split()
review = [ss.stem(word) for word in review if not word in stopwords.words("english")]
review = " ".join(review)
corpus.append(review)
```
